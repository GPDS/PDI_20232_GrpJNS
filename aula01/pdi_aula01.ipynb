```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

cv.__version__
```

```python
img_original = cv.imread('imagens//DSCN5440.JPG')
```

```python
img_gray = cv.cvtColor(img_original, cv.COLOR_RGB2GRAY)

img_red = np.zeros_like(img_original)
img_green = np.zeros_like(img_original)
img_blue = np.zeros_like(img_original)

img_red[:,:,0] = img_original[:,:,0]
img_green[:,:,1] = img_original[:,:,1]
img_blue[:,:,2] = img_original[:,:,2]
```

```python
plt.imshow(cv.cvtColor(img_original, cv.COLOR_BGR2RGB))
plt.axis('off');
```

```python
plt.imshow(img_gray, cmap='gray')
plt.axis('off');
```

```python
plt.imshow(img_red)
plt.axis('off');
```

```python
plt.imshow(img_green)
plt.axis('off');
```

```python
plt.imshow(img_blue)
plt.axis('off');
```
